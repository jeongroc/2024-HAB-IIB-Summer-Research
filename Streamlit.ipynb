```python
#code for the app.py file
from langchain_community.agent_toolkits import create_json_agent, JsonToolkit
from langchain.chat_models import ChatOpenAI
import streamlit as st
import pandas as pd
import json
from langchain.tools.json.tool import JsonSpec

import os
os.environ["OPENAI_API_KEY"] = "your key"

llm = ChatOpenAI(temperature = 0.1, model_name="gpt-4o")

uploaded_file = st.file_uploader("File")
llm_string = st.text_input("Text")

button_clicked = st.button("Run")

if button_clicked:
  if llm_string is not None and uploaded_file is not None:
    data = json.load(uploaded_file)
    spec=JsonSpec(dict_=data)
    toolkit = JsonToolkit(spec=spec)
    agent_executor = create_json_agent(llm, toolkit = toolkit, verbose=True)
    output = agent_executor.run(llm_string)
    st.write(output)
```
